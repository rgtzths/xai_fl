```python
import pandas as pd
import numpy as np
import pathlib

output_path = "./temp_dataset"

output = pathlib.Path(output_path)
output.mkdir(parents=True, exist_ok=True)

identifiers = ["site_id", "mlid", "datetime"]

labels = ["rlf", "1-day-predict", "5-day-predict"]

df_cv = pd.read_csv("preprocessed_dataset/RL_;ABDV/A0BE_time_sentitive_features.csv")

time_sentitive_features = [feature for feature in df_cv.columns if feature not in labels and feature not in identifiers]

X_cv = df_cv[time_sentitive_features].values

y_cv= df_cv[["5-day-predict"]].values

df_train = pd.read_csv("preprocessed_dataset/RL_;ABDV/A0BI_time_sentitive_features.csv")

X_train = df_train[time_sentitive_features].values

y_train = df_train[["5-day-predict"]].values

df_test = pd.read_csv("preprocessed_dataset/RL_;ABDV/A5AB_time_sentitive_features.csv")

X_test = df_test[time_sentitive_features].values

y_test = df_test[["5-day-predict"]].values

df_train_2 = pd.read_csv("preprocessed_dataset/RL_;ABDV/A5AB_time_sentitive_features.csv")

X_train_2 = df_train_2[time_sentitive_features].values

y_train_2 = df_train_2[["5-day-predict"]].values

df_train_3 = pd.read_csv("preprocessed_dataset/RL_;ABDV/A8CQ_time_sentitive_features.csv")

X_train_3 = df_train_2[time_sentitive_features].values

y_train_3 = df_train_2[["5-day-predict"]].values

np.savetxt(output/"x_train_subset_1.csv", X_train, delimiter=",", fmt="%d")
np.savetxt(output/"y_train_subset_1.csv", y_train, delimiter=",", fmt="%d")
np.savetxt(output/"x_train_subset_2.csv", X_train_2, delimiter=",", fmt="%d")
np.savetxt(output/"y_train_subset_2.csv", y_train_2, delimiter=",", fmt="%d")
np.savetxt(output/"x_train_subset_3.csv", X_train_3, delimiter=",", fmt="%d")
np.savetxt(output/"y_train_subset_3.csv", y_train_3, delimiter=",", fmt="%d")

np.savetxt(output/"x_cv.csv", X_cv, delimiter=",", fmt="%d")
np.savetxt(output/"x_test.csv", X_test, delimiter=",", fmt="%d")
np.savetxt(output/"y_cv.csv", y_cv, delimiter=",", fmt="%d")
np.savetxt(output/"y_test.csv", y_test, delimiter=",", fmt="%d")
```
